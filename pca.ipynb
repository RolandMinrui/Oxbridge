```python
from pca.pca import *
import pandas as pd

device = "cuda:0"
model_name_or_path = "sentence-transformers/all-mpnet-base-v2"
train_dataset = []
train_sources = []
train_labels = []

dataset_map = {
    "FPB": {"path": "data/sentimental_analysis/FPB/test.csv", "type": "FPB"},
    "FIQASA": {"path": "data/sentimental_analysis/FIQASA/train.csv", "type": "FIQASA"},
    "OXBRIDGE": {"path": "data/sentimental_analysis/OXBRIDGE/test.csv", "type": "OXBRIDGE"},
}

for dataset_name in dataset_map:
    # load dataset
    df = pd.read_csv(dataset_map[dataset_name]['path']).sample(n=400, random_state=42)
    for i, row in df.iterrows():
        train_dataset.append(row['sentence'])
        train_sources.append(dataset_name)
        train_labels.append(row['label'])
```

```python
engine = Data_Projection(model_name_or_path, device, train_dataset, train_source=train_sources)
```
